```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                        n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
```

```python
def step(z):
      return 1 if z>0 else 0
```

```python
def perceptron(X, y):
      X = np.insert(X, 0, 1, axis=1) # adds 1 as x0 in front of each row
# of X array for calculation of bias upon multiplying with the weights
      weights = np.ones(X.shape[1]) # array of 1's of no of x column [(1,1,1)]as w1, w2, w3
      lr = 0.1

      for i in range(1000):
            j = np.random.randint(0, 100) # selecting a row randomly
            y_hat = step(np.dot(X[j], weights)) # dot product of weights and selected row
            weights = weights + lr * (y[j] - y_hat) * X[j]

      return weights[0], weights[1:] # bias and weights
```

```python
intercept_, coef = perceptron(X,y)
```

```python
print(coef)
print(intercept_)
```

```python
m = -(coef[0]/coef[1])
b = -(intercept_/coef[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```
