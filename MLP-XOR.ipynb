```python
def step_function(x):
      return 1 if x>=0 else 0

def perceptron(x1,x2,w1,w2,b):
      return step_function(w1*x1 + w2*x2 + b)

def OR(x1, x2):
      return perceptron(x1, x2, w1 = 1, w2 = 1, b = -0.5)

def NAND(x1, x2):
      return perceptron(x1, x2, w1 = -1, w2 = -1, b = 1.5)

def AND(x1, x2):
      return perceptron(x1, x2, w1 = 1, w2 = 1, b = - 1.5)

def XOR(x1, x2):
      h1 = OR(x1, x2)
      h2 = NAND(x1, x2)
      return AND(h1, h2)

for x1 in [0,1]:
      for x2 in [0,1]:
            print(x1, x2, "->", XOR(x1, x2))
```
